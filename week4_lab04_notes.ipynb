```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_excel('govt_camp_2013_01_to_2016_03.xlsx')
df['DATE'] = pd.to_datetime(df['DATE'], format='%Y%m%d')
df.insert(0, 'year', pd.DatetimeIndex(df['DATE']).year)
df.insert(1, 'month', pd.DatetimeIndex(df['DATE']).month)
df.insert(2, 'day', pd.DatetimeIndex(df['DATE']).day)
df.drop(['DATE'], axis=1, inplace=True)
df.drop(['STATION','STATION_NAME'], axis=1, inplace=True)
print(df.shape)
df.head()
```

```python
wd = np.array(df)
```

```python
# index out a year using np.where()
locs2013t = np.where(wd[:,0]==2013) # returns locations in a tuple
locs2013 = locs2013t[0]
precip2013 = wd[locs2013,3]

# index out a year using a boolean mask
mask = wd[:,0]==2014    # which rows have 2014 in the first column?
precip2014 = wd[mask,3] # in those rows, pull out the thrid column?


fig, (ax1, ax2, ax3, ax4) = plt.subplots(4, figsize=(9,9))

ax1.plot(precip2013)
ax2.plot(precip2014)
ax3.plot()
ax4.plot()


plt.tight_layout()
plt.show()
```
