## Week 3 Plotting Notes

General Outline to create a Matplotlib plot in Python:

1. Imports
2. Define Data
3. Plot the Data
4. Add customization
5. Show the plot

```python
# 1. Imports
import numpy as np
import matplotlib.pyplot as plt
# if using a Jupyter notebook, include:
%matplotlib inline
```

```python
# 2. Define Data

x = [1,2,3,4]
y = [2,5,8,13]
```

```python
# 3. Plot the data

plt.plot(x,y)

# 4. Add customization

plt.xlabel('x-axis label')
plt.xticks([1.0,1.5,2.0,2.5])
plt.xlim([1.0, 2.5])
plt.ylabel('y-axis label')
plt.title('First Line Plot')

# 5. Show the plot

plt.show()
```
