## A First Simple Plot

```python
# 1. Imports

import numpy as np
import matplotlib.pyplot as plt
# if using a Jupyter notebook, include:
%matplotlib inline
```

```python
# 2. Define Data

x = [1,2,3,4]
y = [2,6,8,13]
```

```python
# 3. Plot the data

plt.plot(x,y)

# 4. Add customization

plt.xlabel('x-axis label')
plt.ylabel('y-axis label')
plt.title('First Line Plot')

# 5. Show the plot

plt.show()
```

## Object-Oriented Plotting

Greater customization is possible when object-oriented plotting is used. Each part of the plot: the figure, axes, lines, tick marks, etc are objects. These objects can be modified with methods and their attributes can be set

```python
# 1. Imports

import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

# 2. Data

x = [1,2,3,4]
y = [2,6,8,13]

# 3. Plot the data

fig, ax = plt.subplots()  # create an axis object, ax
ax.plot(x,y)              # build a plot on the ax object

# 4. Customize the plot
ax.set_xlabel('x-axis label')  # set the ax object attributes
ax.set_ylabel('y-axis label')
ax.set_title('Object-oriented line plot')

# 5. Show

plt.show()
```

## A line plot

```python
#1. Imports
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

#2. Data
x = np.arange(0,np.pi*4,0.1)
y = np.sin(x)

#3. Build Plot
fig, ax = plt.subplots()
ax.plot(x,y)

#4. Customize
ax.set_title('Sine of x from 0 to 4pi')
ax.set_xlabel('x from 0 to 4pi')
ax.set_ylabel('sin(x)')

#5. Show Plot
plt.show()
```

## Multi-line plot

```python
# Imports
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

# Data
x = np.arange(0,np.pi*4,0.1)
y1 = np.sin(x)
y2 = np.sin(x/2)
y3 = 0.5*np.sin(x)

# Plot
fig, ax = plt.subplots()
ax.plot(x,y1,'r')
ax.plot(x,y2,'k')
ax.plot(x,y3,'g')

# Customize
ax.set_title('Multi-line plot')
ax.set_xlabel('x from 0 to 4pi')
ax.set_ylabel('f(x)')
ax.legend(['sin(x)','sin(x/3)','0.5*sin(x)'])

# Show the plot and save
plt.savefig('multi-line_plot.png', dpi=72)
plt.show()
```

## Pie Chart

```python
# imports already complete

# data
os = ['Windows','MacOS','Linux','Chromebook or Tablet']
sizes = [17,6,2,5]

# plot
fig, ax = plt.subplots()
ax.pie(sizes,labels=os, autopct='%1.1f%%')

# customize
ax.axis('equal')  # must set axis 'equal' or pie chart will be an oval
ax.set_title('OS used by ENGR114 Students and Instructor')

# show plot
plt.show()
```

## Bar Chart

```python
# imports already done

# data
os = ['Windows','MacOS','Linux','Chromebook or Tablet']
sizes = [17,6,2,5]
x_pos = [1,2,3,4]

# plot
fig, ax = plt.subplots()
ax.bar(x_pos, sizes, alpha=0.5, align='center')

# customize
ax.set_xticks(x_pos)
ax.set_xticklabels(os)
ax.yaxis.grid(True)
ax.set_title('OS of ENGR114 Students')
ax.set_xlabel('OS used by students')
ax.set_ylabel('Number of students')

# show, save the plot
plt.savefig('bar_plot.png', dpi=72)
plt.show()
```

## Histogram

```python
# imports
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

# data
x = np.random.randn(200)

# plot
fig, ax = plt.subplots()
ax.hist(x, 20) # 20 bins or 20 bars

# customization
ax.set_title('Histogram of 200 normally distributed random numbers')
ax.set_xlabel('Number')
ax.set_ylabel('Frequency')

# show
plt.show()
```
