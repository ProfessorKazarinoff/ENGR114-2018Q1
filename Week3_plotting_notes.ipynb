```python
#1. Imports
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

#2. Data
x = np.arange(0,np.pi*4,0.1)
y = np.sin(x)

#3. Build Plot

fig, ax = plt.subplots()
ax.plot(x,y)

#4. Customize
ax.set_title('Sine of x from 0 to 4pi')
ax.set_xlabel('x from 0 to 4pi')
ax.set_ylabel('sin(x)')

#5. Show Plot
plt.show()
```

```python
# multipline plot

import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

# Data
x = np.arange(0,np.pi*4,0.1)
y1 = np.sin(x)
y2 = np.sin(x/2)
y3 = 0.5*np.sin(x)

# Plot
fig, ax = plt.subplots()
ax.plot(x,y1,'r')
ax.plot(x,y2,'k')
ax.plot(x,y3,'g')

# Customize
ax.set_title('Multi-line plot')
ax.set_xlabel('x from 0 to 4pi')
ax.set_ylabel('f(x)')
ax.legend(['sin(x)','sin(x/3)','0.5*sin(x)'])

# Show the plot and save
plt.savefig('multi-line_plot.png', dpi=72)
plt.show()
```

```python
# Pie Chart

# imports already complete

# data
os = ['Windows','MacOS','Linux','Chromebook or Tablet']
sizes = [17,6,2,5]

# plot
fig, ax = plt.subplots()
ax.pie(sizes,labels=os, autopct='%1.1f%%')

# customize
ax.axis('equal')
ax.set_title('OS used by ENGR114 Students and Instructor')

# show plot
plt.show()
```

```python
# bar chart

# imports already done

# data
os = ['Windows','MacOS','Linux','Chromebook or Tablet']
sizes = [17,6,2,5]
x_pos = [1,2,3,4]

# plot
fig, ax = plt.subplots()
ax.bar(x_pos, sizes, alpha=0.5, align='center')

# customize
ax.set_xticks(x_pos)
ax.set_xticklabels(os)
ax.yaxis.grid(True)
ax.set_title('OS of ENGR114 Students')
ax.set_xlabel('OS used by students')
ax.set_ylabel('Number of students')


# show, save the plot
plt.savefig('bar_plot.png', dpi=72)
plt.show()
```

```python
# Histogram

# imports
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

# data
x = np.random.randn(200)

# plot
fig, ax = plt.subplots()
ax.hist(x, 20) #20 bins or 20 bars

# customization
ax.set_title('Histogram of 200 normally distributed random numbers')
ax.set_xlabel('Number')
ax.set_ylabel('Frequency')

# show
plt.show()
```

```python
# 4 subplots in 1 figure

# imports done

# data
x = np.arange(0.01,5,0.1)
y = 0.5*np.exp(x)

# plots
fig, ( (ax1, ax2) , (ax3,ax4) ) = plt.subplots(2,2)
ax1.plot(x,y)
ax1.set_title('linear x, linear y')

ax2.semilogy(x,y)
ax2.set_title('linear x, log y')

ax3.semilogx(x,y)
ax3.set_title('log x, linear y')

ax4.loglog(x,y)
ax4.set_title('log x, log y')

#show the plot
fig.tight_layout()
plt.savefig('4_subplots.png',dpi=72)
plt.show()

```
